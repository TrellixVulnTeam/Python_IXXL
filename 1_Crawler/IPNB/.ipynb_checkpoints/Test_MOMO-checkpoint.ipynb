```python
#用Network找到有商品資料的url
#因為用網址連不到，所以加上headers-user-agent
import requests
url = "https://m.momoshop.com.tw/search.momo?_advFirst=N&_advCp=N&curPage=1&searchType=1&cateLevel=-1&ent=k&searchKeyword=genius&_advThreeHours=N&_isFuzzy=0&_imgSH=fourCardType"
headers = {"user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Mobile Safari/537.36"}
res = requests.get(url, headers=headers)
res
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'lxml')
soup
```

```python
soup.select("p.prdName")
```

```python
#到網頁inspect查看樹狀標籤，方便看出要的資料在哪，和怎麼定位出來

goods = []
rows = soup.select("li.goodsItemLi")

for row in rows:
    
    name = row.select_one("p.prdName").text.strip()
    price = row.select_one("b.price").text.strip()
    prdEvent = row.select_one("p.prdEvent").text.strip()
    site = row.select_one("a")["href"]
    
    good = [name, price, prdEvent, site]
    goods.append(good)
goods
```

```python
import pandas as pd
df = pd.DataFrame(goods)

df.columns = ['name', 'price', 'prdEvent', 'site']  #改欄位名稱
df
```
