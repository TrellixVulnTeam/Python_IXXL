# 日報今日焦點

```python
import requests
from bs4 import BeautifulSoup


APPLE_DAILY_HEADLINE = 'http://www.appledaily.com.tw/appledaily/hotdaily/headline'
LTN_NEWS = 'https://news.ltn.com.tw/'


def main():
    print('蘋果今日焦點')
    resp = requests.get(APPLE_DAILY_HEADLINE).text
    soup = BeautifulSoup(resp, 'html5lib')
    for headline in soup.find('ul', 'all').find_all('li'):
        print(
            headline.find('div', 'aht_title_num').text, 
            #前十大新聞標籤的class為"aht_title_num atopred"，20~30為"aht_title_num"
            #用前者的話，到第11項會出現AttributeError: 'NoneType' object has no attribute 'text'
            #用後者的話都跑的出來，表示只要前面一樣就好，後面有空白的話，空白之後的省略也可
            headline.find('div', 'aht_title').text,
            headline.find('div', 'aht_title_pv').text
        )
    print('---------')
    print('自由今日焦點')
    resp = requests.get(LTN_NEWS).text
    soup = BeautifulSoup(resp, 'html5lib')
    for lte_news in soup.find('ul','list').find_all('li'):
        print(lte_news.find('span', 'title').text)


if __name__ == '__main__':
    main()
```
