```python
import requests
from bs4 import BeautifulSoup
import time
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://www.amazon.com/s?k={0}&page={1}&qid=1583476177&ref=sr_pg_{1}'  #{0}:format後的第一個位置、{1}:第二個
```

```python
def get_urls(url, query, start_page, end_page): #建立可自行輸入"查詢值"和"start~end_page"的url清單
    urls = []
    for page in range(start_page, end_page+1):
        urls.append(url.format(query, page))    #query帶入url的{0}、page帶入{1}
    return urls
```

```python
def get_resource(url):
    headers = {"user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Mobile Safari/537.36"}
    res = requests.get(url, headers=headers)
    if res.status_code == requests.codes.ok:
        print("請求成功")
        return BeautifulSoup(res.text, "lxml")
    else:
        return print("請求失敗")
```

```python
def get_goods(soup):
    goods = []
    rows = soup.find("div", "sg-col-inner").find_all("div", "s-include-content-margin s-border-bottom")
    for row in rows:
        try:
            name = row.find("span", "a-size-base a-color-base a-text-normal").text
        except:
            None
    
        good = [name]
        goods.append(good)

    return goods
```

```python
def bot(urls):
    all_goods = [["品名"]]  #巢狀清單
    
    page = 1
    for url in urls:
        print("抓取: 第" + str(page) + "頁 網路資料中...")
        page = page + 1
        soup = get_resource(url)
        goods = get_goods(soup)
        all_goods = all_goods + goods
        print("等待3秒鐘...")
        if soup.find("li", class_="a-disabled a-last"):
            break   #已經沒有下一頁
        time.sleep(3)
    return all_goods
```

```python
if __name__ == "__main__":
#get_urls
    urls = get_urls(url, 'logitech', 1, 1)
    print(urls)
#get_resource
    goods = bot(urls)
    df = pd.DataFrame(goods)
    df.columns = ['name']  #改欄位名稱
    print(df)
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml') #常用的解析器有三種："lxml","html5lib","html.parser"，官方建議使用解析速度較快的"lxml"
soup
```

```python
tags = soup.find('div', attrs={'class':'_1HmYoV _35HD7C', 'style':'flex-grow:1;overflow:auto'}).find_all('div', '_3liAhj')

goods = []
for tag in tags:
    try:
        name = tag.find('a', '_2cLu-l').text
    except:
        name = None
    try:
        star = tag.find('div', 'hGSR34').text
    except:
        star = None
    try:
        reviews = tag.find('span', '_38sUEc').text
    except:
        reviews = None
    try:
        price = tag.find('div', '_1vC4OE').text
    except:
        price = None
    try:
        Original_price = tag.find('div', '_3auQ3N').text
    except:
        Original_price = None
    try:
        reviews = tag.find('div', 'VGWI6T').text
    except:
        reviews = None

    good = [name, star, reviews, price, Original_price, reviews]
    goods.append(good)

df = pd.DataFrame(goods)
df.columns = [['name', 'star', 'reviews', 'price', 'Original_price', 'reviews']]
df
```

```python
import requests
from bs4 import BeautifulSoup
import time
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


url_no_pg = "https://www.amazon.com.br/gp/bestsellers/computers/16364917011/ref=zg_bs_nav_computers_4_16364919011" #巴西
URL = url_no_pg +"&pg={0}"  #加入換頁字串

       
def generate_urls(url, start_page, end_page): #使用參數基底URL、開始和結束頁數來建立URL清單
    urls = []   #爬蟲主程式建立的目標網址清單
    for page in range(start_page, end_page+1):
        urls.append(url.format(page)) #format會讓{帶括號}裡的東西格式化
    return urls

if __name__ == "__main__":
    urls = generate_urls(URL, 1, 3)  #得到1~3頁url的
    print(urls)
```
