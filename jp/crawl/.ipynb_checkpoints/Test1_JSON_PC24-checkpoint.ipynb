```python
#用Network找到有商品資料的url
import requests
res = requests.get("https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=%E6%98%86%E7%9B%88&page=1&sort=sale/dc")
res
```

```python
res.text  #發現資料為json形式
```

```python
js = res.json()   #直接轉成json就好(不用用美湯)
js
```

```python
import pandas
df = pandas.DataFrame(js['prods'])
df1 = df[['name', 'describe', 'price', 'originPrice']]
df1
```

```python
df2 = df1.sort_values('price', ascending=False)
df2
```

```python
#抓取更多頁
import requests
products = []
for i in range(1,6):
    res = requests.get(f'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=%E6%98%86%E7%9B%88&page={i}&sort=sale/dc')
    js = res.json()
    df = pandas.DataFrame(js['prods'])
    df1 = df[['name', 'describe', 'price', 'originPrice']]
    products.append(df1)
df1
```

```python
#將每頁合併起來
df_all = pandas.concat(products)
df_all
```
