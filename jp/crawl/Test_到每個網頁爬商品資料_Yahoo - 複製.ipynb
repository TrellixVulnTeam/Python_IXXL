```python
import pandas as pd
import re, time, requests
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
```

# 爬取點入網址

```python
def FindLinks(url):
    linklist = []
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')

    links = soup.find_all('a', {'class':'BaseGridItem__content___3LORP'})
    linklist = []
    for link in links:
        k = link.get('href')
        linklist.append(k)
    return linklist
```

# 爬取點入分頁資料

```python
def GetGoods(response):
    try:
            soup2 = BeautifulSoup(response.text)
            df = pd.DataFrame(data = [{'TITLE':soup2.find('span', attrs={'itemprop':'name'}).text,
                                       'Price':soup2.find('div', attrs={'class':'points enlarge-offset'}).find('del').text,
                                       'Activity Price':soup2.find('span',attrs={'class':'price'}).text,
                                       'reviews':soup2.find('a',attrs={'class':'Grid-U Mstart-4 rating-number'}).text
                                      }],
                                       columns = ['TITLE', 'Price', 'Activity Price', 'reviews']) 
            return df
    
    except AttributeError:
        pass
```

# 開始爬取

```python
urls = 'https://tw.search.mall.yahoo.com/search/mall/product?cid=hp&clv=0&kw=logitech%E7%84%A1%E7%B7%9A%E6%BB%91%E9%BC%A0&p=logitech%E7%84%A1%E7%B7%9A%E6%BB%91%E9%BC%A0'
ndf = []
for i in FindLinks(urls):
    res = requests.get(i)
    g = GetGoods(res)
    ndf.append(g)
    time.sleep(0.5)
ndf
```

# 合併資料
