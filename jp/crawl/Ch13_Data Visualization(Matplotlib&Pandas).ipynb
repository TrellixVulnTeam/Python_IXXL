# Matplot與Pandas的整合
以往我們用matplot畫圖需要將pandas資料丟進matplot畫圖，不過因為現在pandas有整合matplot，<br>
所以可以在pandas的DataFrame下直接畫圖，使用起來更方便，可以再少寫一點程式碼。<br>

```python
#Ch13_1_1
import matplotlib.pyplot as plt

data = [-1, -4.3, 15, 21, 31]
plt.plot(data)  # x軸是 0,1,2,3,4
#您可能想知道為什麼x軸的範圍是0-3，而y軸的範圍是1-4。如果為plot()命令提供單個列表或數組 ，則matplotlib假定它是y值的序列，
#並自動為您生成x值。由於python範圍從0開始，因此默認的x向量的長度與y相同，但從0開始。因此x數據為 [0,1,2,3]。
plt.show()
```

```python
#Ch13_2_3
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("NBA_salary_rankings_2018.csv")
df = df.sort_values("pos")  # 使用位置排序
# 找出位置清單
col = df.drop_duplicates(["pos"])
# 建立各位置薪水的巢狀清單
data = []
for pos in col["pos"].values:
    d = df[(df.pos == pos)]
    data.append(d["salary"].values)
# 繪製箱形圖
plt.boxplot(data)
plt.xticks(range(1,6), col["pos"], rotation=25)
plt.title("Box Plot of NBA Salary") 
plt.show()
```
