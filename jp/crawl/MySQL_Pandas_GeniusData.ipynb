```python
import pymysql
import pandas as pd
import numpy as np

#資料載入
conn = pymysql.connect("localhost", "root", "zx8212290", "genius_data", charset="utf8")
sql = "SELECT * FROM 歷史數據"
df = pd.read_sql(sql, conn)
df
```

```python
# SQL查詢中的 SELECT

a = df[ df['Country'] == 'Taiwan']
a
```

```python
# SELECT多個條件，可以使用 | 或 & 符號連結
b = df[  (df['Year'] == 2014) & (df['Continent'] == '港台')  ]
b
```

```python
# SELECT欄位，用 list 標註變數名稱可以將變數從資料框中選出
c = df[['Order Number','Department','Continent', 'Country']]
c.head(10)
```

```python
# 生成新變數，搭配 apply() 與 lambda 函數將公式應用到每一個觀測值
df['Country_abb'] = df['Country'].apply(lambda x: x[3:5])
df.head(5)
```

```python
# 聚合函數針對欄位計算，實踐 summarise() 函數的功能
df[  (df['Year']==2019) & (df['Country']=='Taiwan')  ]  [['Quantity Shipped']].sum()
```

```python
df[  (df['Year']==2019) & (df['Country']=='Taiwan')  ]  [['Quantity Shipped', 'Foreign Extended Price']].mean()
```

```python
# groupby方法，順序是：SELECT.groupby[數值].聚合函數
df[ (df['Year']==2019) & (df['Department']=='SF1') ].groupby(by = 'Country')[['Quantity Shipped', 'Foreign Extended Price']].sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#折線圖

df_RY = df[['Year', 'Foreign Extended Price']].groupby(by = 'Year')['Foreign Extended Price'].sum()
df_RY.plot(kind = 'line', x = 'Year', y = 'Foreign Extended Price', title = 'Revenue vs. Year', legend = False)
plt.show()
```

```python
#折線圖
#將SF1~SF5資料篩選出來並繪製每年的平均業績變化(使用 pivot_table)：

df_SF = df.loc[df['Department'].isin(['SF1', 'SF2', 'SF3', 'SF4', 'SF5'])]  #isin()接受一个列表，判断该列中元素是否在列表中。
df_SF_pivot = df_SF.pivot_table(values = 'Foreign Extended Price', columns = 'Department', index = 'Year', aggfunc=np.sum)
df_SF_pivot.plot(title = 'Year Revenue by SF')
plt.show()
```

```python
#折線圖(上例的第一行也可以省略)
#將SF1~SF5資料篩選出來並繪製每年的平均業績變化(使用 pivot_table)：

df_SF_pivot2 = df.pivot_table(values = 'Foreign Extended Price', columns = 'Department', index = 'Year', aggfunc=np.sum)
df_SF_pivot2.plot(title = 'Year Revenue by SF')
plt.show()
```

```python
#直方圖(橫軸是代入的值，縱軸是次數)
#將 2019 年資料篩選出來並以三個子圖（subplots）繪製銷量、銷額與單價的直方圖：

df_2019 = df[df['Year'] == 2019]
df_2019[['Quantity Shipped','Foreign Extended Price', 'Foreign Unit Price']].hist(bins = 15)
plt.show()
```

```python
#單價的直方圖：
df_2019[['Foreign Unit Price']].plot(kind = 'hist', title = 'Unit Price in 2019', legend = False, bins = 30)
plt.show()
```

```python
#將單價直方圖依照不同部門以不同顏色繪製：
df_UP_SF_pivot = df_2019.pivot_table(values = 'Foreign Unit Price', columns = 'Department', index = 'Country', aggfunc=np.sum)
df_UP_SF_pivot.plot(kind = 'hist', alpha=0.5, bins = 30, title = 'Unit Price by SF')
plt.show()
```

```python
#以盒鬚圖繪製：
df_UP_SF_pivot.plot(kind = 'box', title = 'Unit Price by SF')
plt.show()
```

```python
#長條圖

df_RY = df[['Year', 'Foreign Extended Price']].groupby(by = 'Year')['Foreign Extended Price'].sum()
df_RY.plot(kind = 'bar', x = 'Year', y = 'Foreign Extended Price', title = 'Revenue vs. Year', legend = False)
plt.show()
```

```python
#長條圖2：
df_x = df[df['Year'] == 2019].groupby(by = 'Department')[['Quantity Shipped', 'Foreign Extended Price']].mean()
df_x.plot(kind = 'barh', subplots = True, layout = (1, 2), sharex = False, sharey = True, legend = False)
plt.show()
```
